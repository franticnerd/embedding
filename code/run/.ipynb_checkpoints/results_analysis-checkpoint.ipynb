## Events

```python
import operator
import collections

def get_event_string(score, points, total_weight):
    common_entities = get_frequent_entities(points)
    s = '%.2f %d'% (score, total_weight) + '\n'
    for entity, frequency in common_entities:
        s += entity + ' ' + '%.2f'% frequency + '\n'
    return s

def get_frequent_entities(points):
    counter = collections.Counter()
    for p in points:
        entity, weight = p['entity'], p['weight']
        counter[entity] += weight
    total_sum = sum([counter[entity] for entity in counter])
    for entity in counter:
        counter[entity] /= total_sum
    return counter.most_common(8)

def plot_events():
    for e in events:
        e_id = e[0]
        plt.figure()
        plt.xlim(-74.3, -73.7)
        plt.ylim(40.5, 40.95)
        score, total_weight, center, members = e[1]['score'], e[1]['totalWeight'], e[1]['center'], e[1]['members']
        s = get_event_string(score, members.values(), total_weight)
        plt.text(-74, 40.75, s)
        x = [p['lng'] for p in members.values()]
        y = [p['lat'] for p in members.values()]
        plt.scatter(x, y, s = 100)
        plt.scatter(center ['lng'], center['lat'])

        
# retrieve the top-k events
def load_events(bandwidth, epsilon, K):
    r = col.find_one({"hubseek.parameters.epsilon" : epsilon, "hubseek.parameters.bandwidth" : bandwidth, 
                     "hubseek.parameters.updateMode" : False})
    # a dict of events: <event id, event>
    event_dict = r['hubseek']['events']
    events = []
    for k, v in event_dict.iteritems():
        events.append([k, v])
    events.sort(key=operator.itemgetter(0), reverse=False)
    return events[:K]


events = load_events(0.01, 0.01, 5)
plot_events()
```

## Running Time

### Scalability of HubSeek

```python
bandwidth_time = m.query("experiments", "event", 
                        {"hubseek.parameters.epsilon" : 0.01, "hubseek.parameters.updateMode" : False},
                         ["hubseek.stats.numTweetsHubSeek", "hubseek.stats.timeHubSeekBatch"],
                         ["int32", "float64"])
df = pd.DataFrame(np.matrix(bandwidth_time).transpose()).sort(1)
df
```

```python
plt.xlabel('# Tweets')
plt.ylabel('Time (sec)')
plt.plot(df.iloc[:,0], df.iloc[:,1], 'rs-')
```

### Update Mode v.s. Batch Mode

```python
update_time = m.query("experiments", "event", 
                        {"hubseek.parameters.bandwidth" : 0.01, "hubseek.parameters.updateMode" : True},
                         ["hubseek.stats.numTweetsHubSeek", "hubseek.stats.numTweetsHubSeekDeletion", "hubseek.stats.numTweetsHubSeekInsertion", 
                          "hubseek.stats.timeHubSeekBatch", "hubseek.stats.timeHubSeekDeletion", "hubseek.stats.timeHubSeekInsertion"],
                         ["int32", "int32", "int32", "float64", "float64", "float64"])
df = pd.DataFrame(np.matrix(update_time).transpose()).sort(2)
df
```

```python
plt.xlabel('Update size')
plt.ylabel('Time (sec)')
p1 = plt.plot(df.iloc[:,1] + df.iloc[:,2], df.iloc[:,4] + df.iloc[:,5], 'bo-')
p2 = plt.plot(df.iloc[:,1] + df.iloc[:,2], df.iloc[:,3], 'rs-')
plt.legend((p1[0], p2[0]), ('Batch', 'Online'), numpoints=1, loc = 'best')
```
